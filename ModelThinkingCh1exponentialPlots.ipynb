```python
#Sigmoid
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values from -50 to 50
x_values = np.linspace(-50, 50, 1000)

# Calculate y values using the sigmoid function
y_values = sigmoid(x_values)

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Sigmoid Function', color='blue')
plt.title('Plot of Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Tanh
import numpy as np
import matplotlib.pyplot as plt

# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Generate x values from -50 to 50
x_values = np.linspace(-50, 50, 1000)

# Calculate y values using the tanh function
y_values = tanh(x_values)

# Plot the tanh function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='tanh Function', color='red')
plt.title('Plot of tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()
plt.show()
```
